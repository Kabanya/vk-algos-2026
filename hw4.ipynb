## 1. Восстановление бинарного дерева из массива

Дан массив, представляющий бинарное дерево (например, в виде уровнями — level-order).
Необходимо восстановить структуру бинарного дерева.

Что нужно сделать:
Реализуйте функцию, которая строит дерево по переданному массиву. Объясните, как определяется связь родитель–потомок и какая структура данных используется для хранения узлов.


```python
class TreeNode:
    def __init__(self, val, left=None, right=None):
        self.val, self.left, self.right = val, left, right
    def print(self, level=0):
        print(" " * 4 * level + "-> " + str(self.val))
        if self.left: self.left.print(level + 1)
        if self.right: self.right.print(level + 1)
```

```python
from collections import deque

def build_tree(vals):
    if not vals or vals[0] is None:
        return None

    root = TreeNode(vals[0])
    queue = deque([root])

    for i in range(1, len(vals), 2):   # шагаем по парам (левый, правый)
        node = queue.popleft()

        if vals[i] is not None:
            node.left = TreeNode(vals[i])
            queue.append(node.left)

        if i + 1 < len(vals) and vals[i + 1] is not None:
            node.right = TreeNode(vals[i + 1])
            queue.append(node.right)

    return root
```

```python
arr = [1, 2, 3, 4, 5, None, 6]
tree = build_tree(arr)
print("Построенное дерево из {}:".format(arr))
tree.print()

arr2 = [1, 2, 3, None, 4, None, None]
tree2 = build_tree(arr2)
print("\nПостроенное дерево 2 из {}:".format(arr2))
tree2.print()

arr3 = [1,2,2,3,3,3,3]
tree3 = build_tree(arr3)
print("\nПостроенное дерево 3 из {}:".format(arr3))
tree3.print()
```

### Пояснение

- В level-order массиве узлы расположены по уровням слева направо.
- Очередь сама отслеживает порядок: каждый узел из очереди получает следующую пару значений из массива как своих потомков.
- TreeNode - каждый узел хранит значение и указатели на потомков.
- Очередь -  временная структура только во время построения. Хранит узлы, которым ещё не назначены потомки.

## 2. Симметричное бинарное дерево

Дано бинарное дерево.
Необходимо определить, является ли оно симметричным относительно своей середины.

Что нужно сделать:
Реализуйте рекурсивную или итеративную проверку зеркальности левого и правого поддеревьев. Поясните, какие пары узлов сравниваются и почему алгоритм работает корректно.

```python
def if_tree_is_symmetric(root):
    def mirror(left, right):
        if not left and not right: return True
        if not left or  not right: return False
        return (left.val == right.val
                and mirror(left.left,  right.right)
                and mirror(left.right, right.left))

    return mirror(root.left, root.right)
```

```python
print("\nПроверка на симметричность:")
print("Дерево {}:".format(arr),  "\tсимметрично:", if_tree_is_symmetric(tree))
print("Дерево {}:".format(arr2), "\tсимметрично:", if_tree_is_symmetric(tree2))
print("Дерево {}:".format(arr3), "\t\tсимметрично:", if_tree_is_symmetric(tree3))
```

### Пояснение

mirror(left, right) рекурсивно проверяет, что значения двух узлов совпадают, а затем перекрёстно сравнивает их потомков: внешние края (left.left с right.right) и внутренние (left.right с right.left).

## 3. Поиск минимальной глубины бинарного дерева

Дано бинарное дерево.
Необходимо найти его минимальную глубину (расстояние от корня до ближайшего листа).
Что нужно сделать:
Реализуйте решение с использованием BFS или DFS. Объясните, почему поиск в ширину (BFS) позволяет корректно найти минимальную глубину.

```python
def min_depth(root):
    if not root:
        return 0

    queue = deque([(root, 1)])  # (узел, текущая глубина)

    while queue:
        node, depth = queue.popleft()

        if not node.left and not node.right:  # лист — сразу возвращаем
            return depth

        if node.left:  queue.append((node.left,  depth + 1))
        if node.right: queue.append((node.right, depth + 1))
```

```python
print("\nПроверка на минимальную глубину:")
print("Дерево {}:".format(arr),  "\tминимальная глубина:", min_depth(tree))
print("Дерево {}:".format(arr2), "\tминимальная глубина:", min_depth(tree2))
print("Дерево {}:".format(arr3), "\t\tминимальная глубина:", min_depth(tree3))

arr4 = [1, 2, 3, 4]
tree4 = build_tree(arr4)


```

### Пояснение

- DFS обойдёт ветку 1→2→4 до конца и вернёт глубину 3, потом проверит 1→3 с глубиной 2 — и только тогда найдёт минимум. Придётся обойти всё дерево.
- BFS идёт уровень за уровнем и останавливается на первом встреченном листе — он гарантированно ближайший. Узлы глубже него даже не проверяются.

```python
print("\nДерево {}:".format(arr4), "минимальная глубина:", min_depth(tree4))
tree4.print()
```

## 4. Поиск произведения максимального и минимального элементов

Дано бинарное дерево поиска (BST), представленное в виде массива.
Необходимо найти произведение минимального и максимального значений.
Что нужно сделать:
Восстановите BST (если требуется) и найдите минимальный и максимальный элементы, используя свойства BST. Поясните, почему поиск крайних элементов выполняется за O(h), где h — высота дерева.

```python
def build_bst(vals):
    if not vals or vals[0] is None:
        return None
    root = TreeNode(vals[0])
    queue = deque([root])
    for i in range(1, len(vals), 2):
        node = queue.popleft()
        if vals[i] is not None:
            node.left = TreeNode(vals[i])
            queue.append(node.left)
        if i + 1 < len(vals) and vals[i + 1] is not None:
            node.right = TreeNode(vals[i + 1])
            queue.append(node.right)
    return root

def min_node(node):
    while node.left:
        node = node.left
    return node.val

def max_node(node):
    while node.right:
        node = node.right
    return node.val

def min_max_product(vals):
    root = build_bst(vals)
    mn, mx = min_node(root), max_node(root)
    print(f"min={mn}, max={mx}, product={mn * mx}")
```

```python
min_max_product([8, 3, 10, 1, 6, None, 14])
min_max_product([1, None, 2, None, 3, None, 4])
min_max_product([42])
min_max_product([0, -5, 7, -10, -1, None, 9])
```

### Пояснение

Свойства BST

Минимум — самый левый узел
Максимум — самый правый узел

Идти вглубь всего дерева не нужно.

На каждом шаге делается один выбор — идти левее или правее. Никаких ветвлений, никакого возврата. Количество шагов равно высоте дерева h

## 5. Сравнение двух бинарных деревьев

Даны два бинарных дерева.
Необходимо определить, являются ли они идентичными (по структуре и значениям узлов).

Что нужно сделать:
Реализуйте рекурсивную проверку соответствия узлов обоих деревьев. Объясните, какие условия должны выполняться, чтобы деревья считались одинаковыми.

```python
def is_same(p, q):
    if not p and not q: return True   # оба пусты
    if not p or  not q: return False  # одно пусто
    return (p.val == q.val
            and is_same(p.left,  q.left)
            and is_same(p.right, q.right))


print("\nПроверка на идентичность:")
print("Дерево {} и {}:".format(arr, arr2), "\tидентичны:", is_same(tree, tree))
print("Дерево {} и {}:".format(arr, arr3), "\t\tидентичны:", is_same(tree, tree3))
print("Дерево {} и {}:".format(arr2, arr3), "\t\tидентичны:", is_same(tree2,tree3))
```

### Пояснение

Условия идентичности
Деревья одинаковы, если одновременно:

- оба узла существуют
- их значения совпадают
- левые поддеревья идентичны
- правые поддеревья идентичны
