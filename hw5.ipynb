## Полное бинарное дерево


Реализуйте функцию, которая проверяет, является ли заданное бинарное дерево полным.
Необходимо корректно определить условие заполненности уровней и объяснить выбранный способ обхода (например, BFS).

```python
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def is_complete(root: TreeNode) -> bool:
    if not root:
        return True  # пустое дерево полное по определению
    queue = deque([root])
    must_be_leaf = False  # флаг «ожидаем только листья»

    while queue:
        node = queue.popleft()
        if node.left:
            if must_be_leaf:
                return False  # нарушение!
            queue.append(node.left)
        else:
            must_be_leaf = True  # левого нет → ждём только листьев
        if node.right:
            if must_be_leaf:
                return False  # нарушение!
            queue.append(node.right)
        else:
            must_be_leaf = True  # правого нет → ждём только листьев

    return True
```

```python
tree1 = TreeNode(1)
tree1.left = TreeNode(2, TreeNode(4), TreeNode(5))
tree1.right = TreeNode(3, TreeNode(6))
print("Пример 1:", is_complete(tree1))

tree2 = TreeNode(1)
tree2.left = TreeNode(2, TreeNode(4))  # Правого потомка у 2 нет
tree2.right = TreeNode(3, TreeNode(6))
print("Пример 2:", is_complete(tree2))

tree3 = TreeNode(1)
tree3.left = TreeNode(2, None, TreeNode(5)) # Пропущен левый, но есть правый
tree3.right = TreeNode(3)
print("Пример 3:", is_complete(tree3))

tree4 = None
print("Пример 4:", is_complete(tree4))
```
