# ДЗ 2 — Односвязные списки, строки и два указателя

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def list_to_linked(lst):
    dummy = ListNode()
    cur = dummy
    for v in lst:
        cur.next = ListNode(v)
        cur = cur.next
    return dummy.next


def linked_to_list(head):
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return result
```

## Задача 1. Развернуть односвязный список


```python
def reverse_list(head: ListNode) -> ListNode:
    prev = None
    curr = head
    while curr:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node
    return prev
```

```python
assert linked_to_list(reverse_list(list_to_linked([1, 2, 3, 4, 5]))) == [5, 4, 3, 2, 1]
assert linked_to_list(reverse_list(list_to_linked([1]))) == [1]
assert linked_to_list(reverse_list(list_to_linked([]))) == []
assert linked_to_list(reverse_list(list_to_linked([10, 20]))) == [20, 10]
```

## Задача 2. Найти середину списка


```python
def middle_of_list(head: ListNode) -> ListNode:
    slow = head
    fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow
```

```python
assert middle_of_list(list_to_linked([1, 2, 3, 4, 5])).val == 3
assert middle_of_list(list_to_linked([1, 2, 3, 4, 5, 6])).val == 4
assert middle_of_list(list_to_linked([1])).val == 1
assert middle_of_list(list_to_linked([1, 2])).val == 2
```

## Задача 3. Удалить элемент из односвязного списка

```python
def remove_elements(head: ListNode, val: int) -> ListNode:
    dummy = ListNode(0, head)
    prev = dummy
    curr = head
    while curr:
        if curr.val == val:
            prev.next = curr.next
        else:
            prev = curr
        curr = curr.next
    return dummy.next
```

```python
assert linked_to_list(remove_elements(list_to_linked([1, 2, 6, 3, 4, 5, 6]), 6)) == [1, 2, 3, 4, 5]
assert linked_to_list(remove_elements(list_to_linked([7, 7, 7, 7]), 7)) == []
assert linked_to_list(remove_elements(list_to_linked([1, 2, 3]), 4)) == [1, 2, 3]
assert linked_to_list(remove_elements(list_to_linked([]), 1)) == []
assert linked_to_list(remove_elements(list_to_linked([1]), 1)) == []
```

## Задача 4. Является ли одна строка исходной для другой

```python
def is_subsequence(a: str, b: str) -> bool:
    i = 0
    j = 0
    while i < len(a) and j < len(b):
        if a[i] == b[j]:
            i += 1
        j += 1
    return i == len(a)
```

```python
assert is_subsequence("abc", "aXbYcZ") == True
assert is_subsequence("abc", "abc") == True
assert is_subsequence("axc", "ahbgdc") == False
assert is_subsequence("", "anything") == True
assert is_subsequence("a", "") == False
assert is_subsequence("ace", "abcde") == True
print("Задача 4 — все тесты пройдены ✓")
```

## Задача 5. Метод двух указателей — поиск пары с заданной суммой

```python
def two_sum_sorted(arr: list[int], target: int) -> bool:
    left = 0
    right = len(arr) - 1
    while left < right:
        s = arr[left] + arr[right]
        if s == target:
            return True
        elif s < target:
            left += 1
        else:
            right -= 1
    return False
```

```python
assert two_sum_sorted([1, 2, 3, 4, 6], 6) == True
assert two_sum_sorted([1, 2, 3, 4, 6], 12) == False
assert two_sum_sorted([1, 3], 4) == True
assert two_sum_sorted([1, 3], 5) == False
assert two_sum_sorted([-3, -1, 0, 2, 4, 5], 1) == True
assert two_sum_sorted([], 0) == False
```

## Задача 6. Является ли слово палиндромом?

```python
def is_palindrome(s: str) -> bool:
    left = 0
    right = len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True
```

```python
assert is_palindrome("abba") == True
assert is_palindrome("racecar") == True
assert is_palindrome("hello") == False
assert is_palindrome("a") == True
assert is_palindrome("") == True
assert is_palindrome("ab") == False
assert is_palindrome("abacaba") == True
```

## Задача 7. Метод двух указателей — удаление дубликатов

```python
def remove_duplicates(arr: list[int]) -> int:
    if not arr:
        return 0
    slow = 0
    for fast in range(1, len(arr)):
        if arr[fast] != arr[slow]:
            slow += 1
            arr[slow] = arr[fast]
    return slow + 1
```

```python
a1 = [1, 1, 2]
k1 = remove_duplicates(a1)
assert k1 == 2 and a1[:k1] == [1, 2]

a2 = [0, 0, 1, 1, 1, 2, 2, 3, 3, 4]
k2 = remove_duplicates(a2)
assert k2 == 5 and a2[:k2] == [0, 1, 2, 3, 4]

a3 = [1, 2, 3]
k3 = remove_duplicates(a3)
assert k3 == 3 and a3[:k3] == [1, 2, 3]

a4 = [5, 5, 5, 5]
k4 = remove_duplicates(a4)
assert k4 == 1 and a4[:k4] == [5]

assert remove_duplicates([]) == 0
```

## Задача 8. Слияние двух отсортированных списков

```python
def merge_two_lists(l1: ListNode, l2: ListNode) -> ListNode:
    dummy = ListNode()
    tail = dummy
    while l1 and l2:
        if l1.val <= l2.val:
            tail.next = l1
            l1 = l1.next
        else:
            tail.next = l2
            l2 = l2.next
        tail = tail.next
    tail.next = l1 if l1 else l2
    return dummy.next
```

```python
res = merge_two_lists(list_to_linked([3, 6, 8]), list_to_linked([4, 7, 9, 11]))
assert linked_to_list(res) == [3, 4, 6, 7, 8, 9, 11]
assert linked_to_list(merge_two_lists(list_to_linked([1, 2, 3]), None)) == [1, 2, 3]
assert linked_to_list(merge_two_lists(None, list_to_linked([4, 5]))) == [4, 5]
assert linked_to_list(merge_two_lists(None, None)) == []

res2 = merge_two_lists(list_to_linked([1]), list_to_linked([2, 3, 4]))
assert linked_to_list(res2) == [1, 2, 3, 4]
```
